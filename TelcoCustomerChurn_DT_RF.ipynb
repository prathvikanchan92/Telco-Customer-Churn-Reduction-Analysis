```python
#creating dummy variables
#for MultipleLines,InternetService,contract,payment
telcoDataset = pd.get_dummies(telcoDataset, columns=['MultipleLines','InternetService','Contract','PaymentMethod'])
telcoDataset.shape
```

```python
#rearranging the column
telcoDataFrame = pd.DataFrame(telcoDataset)
#dropping additional dummy variable
telcoData = telcoDataFrame.drop(["customerID",'MultipleLines_No','InternetService_Fiber optic','Contract_Two year','PaymentMethod_Mailed check'],axis=1)
telcoData = telcoDataFrame[['Churn (Yes=1)'] + [colmn for colmn in telcoData.columns if colmn != 'Churn (Yes=1)']]
feature_names=telcoData.columns.tolist()
print(telcoData.shape)
print(feature_names)
```

```python
# Install graphviz
#import sys
#!conda install --yes --prefix "{sys.prefix}" python-graphviz
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#dropping additional redundant dummy variable
X = telcoData.loc[:,'gender':'PaymentMethod_Electronic check']
y = telcoData.loc[:,'Churn (Yes=1)']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.4,random_state=0)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_test_hat = model.predict(X_test)
print("Out-sample accuracy for decision tree is {}".format(accuracy_score
                        (y_test,y_test_hat.round(),normalize=True)*100))
```

```python
feature_names=['Churn (Yes=1)','gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure', 'PhoneService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'MonthlyCharges', 'MultipleLines_No phone service', 'MultipleLines_Yes', 'InternetService_DSL', 'Contract_Month-to-month', 'Contract_One year', 'PaymentMethod_Bank transfer (automatic)', 'PaymentMethod_Credit card (automatic)', 'PaymentMethod_Electronic check']
for feature in zip(feature_names[1:], model.feature_importances_):
    print(feature)
```

Based on the above feature importances we can see that MonthlyCharges,tenure,Contract_month_to_month are important in deciding customer's churn.

```python
#visualizing the decision tree
import graphviz
dot_data = export_graphviz(model, 
                           feature_names=feature_names[1:],  
                           class_names=feature_names[0],  
                           filled=True, rounded=True,  
                           special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
#implementing random forest model
import numpy as np
from sklearn.ensemble import RandomForestClassifier
rfmodel = RandomForestClassifier(n_estimators= 10000,n_jobs=-1, random_state=0)
#training the model
rfmodel.fit(X_train,y_train)
y_test_hat = rfmodel.predict(X_test)
print("accuracy score of the random forest model is {}".format(accuracy_score(y_test,y_test_hat.round(),normalize=True)*100))
```

```python
# Printing the name and gini importance of each feature
for feature in zip(feature_names[1:], rfmodel.feature_importances_):
    print(feature)
```

From the random forest prediction model above, we see that tenure, monthly charges are siginificant features. We can round up Contract_Month-to-month and that is approximately 0.1 which is also a significant feature.
